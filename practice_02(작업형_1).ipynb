```python
## 01.
# 인기동영상 제작횟수가 많은 채널 상위 10개명을 출력하라 (날짜기준, 중복포함)
```

```python
# print(list(df.loc[df['channelId'].isin(df['channelId'].value_counts().head(10).index)]['channelTitle'].unique()))

top = df['channelId'].value_counts().head(10).index

print(list(df.loc[df['channelId'].isin(top)]['channelTitle'].unique()))
```

```python
## 02.
# 논란으로 인기동영상이 된 케이스를 확인하고 싶다. dislikes수가 like 수보다 높은 동영상을 제작한 채널을 모두 출력하라
```

```python
print(list(df[df['likes'] < df['dislikes']]['channelTitle'].unique()))
```

```python
## 03.
# 채널명을 바꾼 케이스가 있는지 확인하고 싶다. channelId의 경우 고유값이므로 이를 통해 채널명을 한번이라도 바꾼 채널의 갯수를 구하여라
```

```python
# change = df[['channelTitle','channelId']].drop_duplicates().channelId.value_counts()
# target = change[change>1]
# print(len(target))

change = df[['channelTitle', 'channelId']].drop_duplicates().channelId.value_counts()
target = change[change>1]
print(len(target))
```

```python
## 04.
# 일요일에 인기있었던 영상들중 가장많은 영상 종류(categoryId)는 무엇인가?
```

```python
df.info()
```

```python
## 05.
# 각 요일별 인기 영상들의 categoryId는 각각 몇개 씩인지 하나의 데이터 프레임으로 표현하라
```

```python
## 05.
# 남성 중 A등급과 D등급의 체지방률 평균의 차이(큰 값에서 작은 값의 차)를 구하여라
```

```python
m_a = df[(df['측정회원성별']=='M') & (df['등급']=='A')]['체지방율 : %'].mean()
m_d = df[(df['측정회원성별']=='M') & (df['등급']=='D')]['체지방율 : %'].mean()

print(abs(m_a - m_d))
```

```python
## 06.
# 여성 중 A등급과 D등급의 체중의 평균의 차이(큰 값에서 작은 값의 차)를 구하여라
```

```python
w_a = df[(df['측정회원성별']=='F') & (df['등급']=='A')]['체중 : kg'].mean()
w_d = df[(df['측정회원성별']=='F') & (df['등급']=='D')]['체중 : kg'].mean()

print(abs(w_a - w_d))
```

```python
## 07.
# bmi는 자신의 몸무게(kg)를 키의 제곱(m)으로 나눈값이다. 데이터의 bmi 를 구한 새로운 컬럼을 만들고 남성의 bmi 평균을 구하여라
```

```python
df['bmi'] = df['체중 : kg'] / (df['신장 : cm']/100) ** 2
m_bmi = df[df['측정회원성별'] == 'M']['bmi'].mean()
print(m_bmi)
```

```python
## 08.
# bmi보다 체지방율이 높은 사람들의 체중평균을 구하여라
```

```python
print(df[df['bmi'] < df['체지방율 : %']]['체중 : kg'].mean())
```

```python
## 09.
# 남성과 여성의 악력 평균의 차이를 구하여라
```

```python
p_m = df[df['측정회원성별'] == 'M']['악력D : kg'].mean()
p_f = df[df['측정회원성별'] == 'F']['악력D : kg'].mean()

print(abs(p_m - p_f))
```

```python
## 10.
# 남성과 여성의 교차윗몸일으키기 횟수의 평균의 차이를 구하여라
```

```python
p_m = df[df['측정회원성별'] == 'M']['교차윗몸일으키기 : 회'].mean()
p_f = df[df['측정회원성별'] == 'F']['교차윗몸일으키기 : 회'].mean()

print(abs(p_m - p_f))
```

## nba 선수 능력치 데이터

```python
pd.set_option('display.max_columns',50)
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/nba/nba.csv",encoding='latin',sep=';')
df.head()
```

```python
## 01.
# Tm 컬럼은 각 팀의 이름을 의미한다. TOR팀의 평균나이를 소수 4째 자리까지 구하여라
```

```python
print(round(df[df['Tm'] == 'TOR']['Age'].mean(), 4))
```

```python
## 02.
# Pos 컬럼은 포지션을 의미한다. 전체 선수 중 최소나이대의 선수들을 필터하고 그들 중 가장 많은 포지션은 무엇인지 확인하라
```

```python
print(df[df['Age'] == df['Age'].min()]['Pos'].value_counts().index[0])
```

```python
## 03.
# 선수들의 이름은 first_name+ 공백 + last_name으로 이루어져 있다. 가장 많은 first_name은 무엇이며 몇 회 발생하는지 확인하라
```

```python

a = []

for i in range(0, len(df)):
  b = df['Player'].iloc[i].split(' ')[0]
  a.append(b)

c = pd.DataFrame(a)

print(c.value_counts().index[0], c.value_counts().iloc[0])

# result= df['Player'].str.split().str[0].str.lower().value_counts().head(1)
# print(result)
```

```python
## 04.
# PTS컬럼은 경기당 평균득점수 이다. 각포지션별로 경기당 평균득점수의 평균을 구하여라
```

```python
print(df.groupby(['Pos'])['PTS'].mean().sort_values())
```

```python
## 05.
# G컬럼은 참여한 경기의 숫자이다. 각 팀별로 가장 높은 경기참여 수를 가진 선수들의 경기 참여 숫자의 평균을 구하여라
```

```python
print(df.groupby(['Tm'])['G'].max().mean())

# result = df.sort_values(['Tm','G']).groupby('Tm').tail(1).G.mean()
# print(result)
```

```python
## 06.
# Tm의 값이 MIA이며 Pos는 C또는 PF인 선수의 MP값의 평균은?
```

```python
print(df[(df['Tm'] == 'MIA') & ((df['Pos'] == 'C') | (df['Pos'] == 'PF'))]['MP'].mean())
```

```python
## 07.
# 전체 데이터중 G의 평균값의 1.5배 이상인 데이터들만 추출했을때 3P값의 평균은?
```

```python
print(df[df['G'] >= 1.5 * df['G'].mean()]['3P'].mean())
```
